#Multiple linear regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#preprocessing

```python
# Load selected dataset (Replace with actual filename)
df = pd.read_csv("/content/selected_features.csv")

# Load dataset
df = pd.read_csv("/content/selected_features.csv")

# Select only numerical columns
numerical_columns = df.select_dtypes(include=['number'])

# Count the number of numerical columns
num_numerical_vars = numerical_columns.shape[1]

print("Numerical Columns:", numerical_columns.columns.tolist())
print("Total Numerical Variables:", num_numerical_vars)
```

```python
# Select only categorical columns (object and boolean types)
categorical_columns = df.select_dtypes(include=['object', 'bool'])

# Count the number of categorical columns
num_categorical_vars = categorical_columns.shape[1]

print("Categorical Columns:", categorical_columns.columns.tolist())
print("Total Categorical Variables:", num_categorical_vars)
```

```python
# Define target and features
target = "Fuel Information.Highway mpg"
features = df.columns[:-1]

# One-Hot Encoding for Categorical Features
df = pd.get_dummies(df, columns=['Engine Information.Driveline', 'Engine Information.Transmission'], drop_first=True)

# Splitting data into input (X) and target (Y)
X = df.drop(columns=[target])
Y = df[target]
```

#Model training and prediction

```python
# Train-Test Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Multiple Linear Regression Model
model = LinearRegression()
model.fit(X_train_scaled, Y_train)

# Predictions
Y_pred = model.predict(X_test_scaled)

# Model Evaluation
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"RÂ² Score: {r2:.2f}")
```

#plotting

```python
# Scatter Plot: Actual vs Predicted Values
plt.figure(figsize=(8, 6))
sns.scatterplot(x=Y_test, y=Y_pred)
plt.xlabel("Actual Highway MPG")
plt.ylabel("Predicted Highway MPG")
plt.title("Actual vs Predicted Highway MPG")
plt.axline([0, 0], [1, 1], color='red', linestyle='dashed')  # Ideal line
plt.show()

# Residual Plot
residuals = Y_test - Y_pred
plt.figure(figsize=(8, 6))
sns.histplot(residuals, bins=30, kde=True)
plt.xlabel("Residuals (Actual - Predicted)")
plt.ylabel("Frequency")
plt.title("Residuals Distribution")
plt.show()
```
